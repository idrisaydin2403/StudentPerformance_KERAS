```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\İdris Aydın\\Desktop\\student_performance_data.csv")
```

```python
df
```

```python
df = df.drop(columns = ['StudentID', 'Gender', 'Age', 'Major'])
```

```python
df
```

```python
pd.set_option('future.no_silent_downcasting', True)
```

```python
df['PartTimeJob'] = df['PartTimeJob'].replace({'Yes': 1, 'No': 0})
df['ExtraCurricularActivities'] = df['ExtraCurricularActivities'].replace({'Yes': 1, 'No': 0})
```

```python
df['PartTimeJob'] = df['PartTimeJob'].astype(int)
df['ExtraCurricularActivities'] = df['ExtraCurricularActivities'].astype(int)
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
feature_scaler = MinMaxScaler()
target_scaler = MinMaxScaler()
```

```python
features = df[['StudyHoursPerWeek', 'AttendanceRate']]
target = df[['GPA']]
```

```python
scaled_features = feature_scaler.fit_transform(features)
scaled_target = target_scaler.fit_transform(target.values.reshape(-1, 1))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_features, scaled_target, test_size=0.33, random_state = 42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential(
    [
        Dense(64, activation = 'relu'),
        Dense(64, activation = 'relu'),
        Dense(1)
    ]
)
```

```python
model.compile(optimizer='adam', loss='mse', metrics = ['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs = 100, batch_size=8)
```

```python
test_loss, test_accuracy = model.evaluate(x_test, y_test)
```

```python
prediction = model.predict(x_test)
print(f"Prediction:{prediction}")
```

```python
new_data = {
    'StudyHoursPerWeek': [40],
    'AttendanceRate': [75.8]
}

new_df = pd.DataFrame(new_data)

new_scaled_features = feature_scaler.fit_transform(new_df)
```

```python
prediction_scaled = model.predict(new_scaled_features)
```
